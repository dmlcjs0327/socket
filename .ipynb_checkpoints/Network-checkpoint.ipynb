# 통신(Communication)

- <font color=blue>**데이터를 주고받는 모든 행위 및 체계**</font>
- **비전자 통신**: 봉화, 파발, 등대, 편지, 신문, ... etc
- **전자 통신**: 인터넷, 전화, TV, ...etc

#### ※데이터와 정보의 차이
- **데이터(data)**: 일어난 사실, 값, 자료 등의 총칭
- **정보(information)**: 데이터를 가공하여 유용한 값으로 바꾼 것

---
# 교환 기술(Switching Technology)



- **회선 교환**: 데이터가 고정적인 회선을 통해 전달되는 방식  


- **축적 후 교환**: 데이터가 각 노드를 경유하면서 (축적되어) 전달되는 방식  

  - **메세지 교환**: 데이터 전체가 통째로 전달되는 방식  
  
  - **패킷 교환**: 데이터가 패킷으로 분할 후 각각 다른 경로를 거쳐 전달되는 방식    
  
    - **데이터그램 방식**: 각 패킷이 독립적으로 매 순간 최적의 경로를 선택하여 전달되는 방식 (매번 경로를 새로 탐색)
    
    - **가상회선 방식**: 패킷들이 정해진 경로를 따라 전달되는 방식 (경로를 처음에만 한 번 탐색)

# 네트워크(Network)

- <font color=blue>**여러 개체(node)가 그물처럼 연결된 상태**</font> 또는 <font color=blue>**각 개체가 데이터를 주고 받는 경로**</font> = 연결망
- 일반적으로 여러 컴퓨터가 연결되어 데이터를 주고 받는 경로를 의미하며, 이는 정확히는 **통신망(Telecommunication Network)** 이라고 함
- 그물(Net)을 짜는 일(work)을 의미하는 명사에서, 연결망이란 의미로 확장된 단어
- **네트워크의 평가**: 정보전달량, 지연시간, 신뢰성, 보안성을 통해 평가

---
# 서버(Server) & 클라이언트(Client)
- **노드(Node)**: 네트워크 상의 모든 장치


- **호스트(Host)**: 네트워크를 사용하려는 노드  
※<font color=red>**호스트가 아닌 노드: 인터넷 공유기, 스위치 등은 데이터가 거쳐가는 장소일 뿐 직접적으로 이용하려는 장치는 아님**</font>


- **클라이언트(Client)**: 요청하는 호스트  


- **서버(Server)**: 요청에 응답할 수 있는 호스트  
※<font color=red>**주의: 서버를 그냥 호스트라고 하는 경우도 많음**</font>



---
# 네트워크 종류 (Types of Network)


  
  
<center> 규모에 따라 분류해놓은 통신망 비교도 </center>

### - 광역 통신망 (Wide Area Network / WAN)
- <font color=blue>**가장 상위 단위의 통신망**</font>
- 전국에 깔린 **전화선**, **인터넷** 서비스를 위한 회선, 기업의 인트라넷에서 본사/지사를 전용선으로 연결한 경우 등 

### - 도시권 통신망 (Metropolitan Area Network / MAN)
- <font color=blue>**도시 각각에 설치된 통신망**</font>
- 각 MAN은 다른 도시의 MAN과 WAN으로 연결
- 통신망 개념에 처음부터 있던 개념이 아니며, 휴대전화의 보급을 위해 각 도시별 **기지국**을 세우게 되면서 이를 가리키기 위해 생김

### - 근거리 통신망 (Local Area Network / LAN)
- <font color=blue>**하나의 사무실 혹은 주택 내의 통신망**</font>
- WAN 또는 MAN과 연결되며, 통신망의 대부분을 차지함

### - 개인 통신망 (Personal Area Network / PAN)
- <font color=blue>**한 사람이 사용하는 다수의 장비 간 통신망**</font>
- 블루투스, 스마트폰 핫스팟 등

---  
# 대역폭 (Bandwidth)

- <font color=blue>**네트워크에서 이용할 수 있는 신호의 최고 주파수와 최저 주파수의 차**</font>
- 주파수이므로 **단위는 초당 진동횟수(Hz: 헤르츠)**
- 대역폭이 클수록 많은 주파수 영역을 사용할 수 있으므로, 한 번에 여러 주파수 영역을 사용하여 데이터를 전송할 수 있음  
 **※주의**: <font color=red>**대역폭과 전송속도가 항상 비례관계는 아님**</font>  
 (대역폭이 좁을 때 더 전송속도가 올라가는 경우도 있고, 데이터 변조를 통해 압축하여 보낼 경우 대역폭과 관계없이 많이 보낼 수 있음) 

#### ※데이터 전송속도의 단위: 초당 전송비트(bps: 비트레이트)

---
# 토폴로지 (Topology)

- [일반] 외형적인 연결 모양
- [수학] 위상
- <font color=blue>**[통신] 네트워크의 구성 형태**</font>

### - 그물형(Mesh Topology)
  - <font color=blue>**그물처럼 각 노드가 서로 연결된 형태**</font> (만약 모든 노드끼리 일대일로 연결된 경우, **full mesh** 혹은 **full connected** 라고 함)
  - 통신선로의 총길이가 가장 긴 네트워크
  - 초기 데이터 통신 네트워크의 전형적인 형태
  - 한 노드가 손상되더라도 우회가능하므로 안정성이 높음 / 단, 무한순환이 발생할 수 있으므로 경로 탐색하는 방식(Routing)이 중요
  
  
### - 트리형(Tree Topology)
  - <font color=blue>**계층구조를 가지는 연결 형태**</font>
  - 주로 WAN에서 사용 (고속망으로 따온 회선을 최종 단말로 분배하는 식)
  
  
### - 고리형(Ring Topology / Loop)
  - <font color=blue>**각 링크가 단방향이어서 데이터가 한 방향으로만 전송되는 연결 형태**</font>
  - 원형 선로를 따라 이동하는 토큰이 있으며, 토큰이 위치한 노드가 데이터를 송수신 (**Token Passing**)
  - 토큰이 위치한 노드가 손상되는 경우 전체 네트워크가 마비
  
  
### - 별형(Star Topology / Hub and Spoke)
  - <font color=blue>**중앙 집중식 연결 형태**</font>
  - 통신망의 제어가 간단하고 구축 비용도 저렴
  - 이용자의 스테이션은 중앙의 교환장비와 점대점(PTP)으로 연결
  - 중앙 노드가 손상되는 경우 전체 네트워크가 마비 (단 그 외의 다른 노드는 손상되더라도 해당 노드에만 문제 있고 전체엔 문제 없음)
    
    
### - 버스형(Bus Topology)
  - <font color=blue>**모든 노드들은 간선을 공유하며,T자형(Tap)으로 연결된 형태**</font>
  - 노드가 추가될 때마다 추가되는 회선이 적기에 구축비용은 저렴하다, 중앙선에 데이터가 몰리는 병목현상 발생

---
# 프로토콜(Protocol)

- 외규규약: 국가 간 원할한 교류를 위해 만들어진 규약(약속, 규율, 국제법 등)
- <font color=blue>**통신규약: 컴퓨터들 간 원할한 통신을 위해 만들어진 규약(신호처리 방법, 오류처리 방법, 암호, 인증사항, 주소 등)**</font>  
- 여러 프로토콜의 모음은 **프로토콜 스위트(Protocol Suite)**라고 하기도 함 / ex) 인터넷 프로토콜 스위트(Internet Protocol Suite)

※ 원할한 통신을 위해선 프로토콜이 통일되어 있어야 하므로, 전세계적으로 통용되는 **국제 표준 통신규약**이 존재 (UN 산하의 ITU에서 만든 규약)

※프로토콜은 말 그래도 '규약'이란 의미이므로, 네트워크의 여러 부분에 대한 프로토콜들이 각각 존재함  
  (인터넷에 대한 프로토콜, 장치 간 전송에 대한 프로토콜 등 전송 종류에 따라 각각 별도의 프로토콜들이 존재한다는 의미)

※ 모든 신호는 비트(0과 1)로 보내지기 때문에, 프로토콜을 알지 못하면 전송된 데이터를 해석하는 것 자체가 어려울 것임

---
# OSI 모형 (Open Systems Interconnection Reference Model)

- <font color=blue>**컴퓨터 네트워크 프로토콜의 디자인과 통신을 7개 계층으로 나눈 모델**</font> (국제표준화기구(ISO)에서 개발)
- 이상이 발생 시, 전체 구조를 수정할 필요 없이 이상이 생긴 단계에 해당하는 하드웨어/소프트웨어만 수정하면 됨
- 7계층이 사람과 소통하는 계층이고 1계층은 순수히 기계적인 계층이므로, 보통 7계층에서 1계층까지 하향식으로 설명
- 사람/7/6/5/4/3/2/1/기계 -> 기계/1/2/3/4/5/6/7/사람 순으로 통신이 이뤄지기 때문에, 사용자들을 보통 **양끝단(end-to-end) 사용자**라고 함




### -7계층: 응용 계층(Application Layer)
- <font color=blue>**응용 프로세스(사용자)와 직접 연관되어, 일반적인 응용 서비스를 수행하는 계층**</font>
- **프로토콜**: HTTP, FTP, SMTP, POP3, IMAP, Telnet 
- **하는 일**: 사용자가 입력한 데이터를 프로토콜에 맞게 수정 후 6계층으로 전달/6계층에서 온 데이터가 프로토콜에 맞는지 검사


### -6계층: 표현 계층(Presentation Layer)
- <font color=blue>**데이터 표현이 상이한 응용 프로세스의 독립성을 제공하고 전송 데이터를 암호화 하는 계층**</font>
- **하는 일**: 데이터의 인코딩/디코딩, 암호화/복호화, 압축/압축 해제
  

### -5계층: 세션 계층(Session Layer)
- <font color=blue>**양 끝단의 사용자들이 통신을 관리하기 위한 방법을 제공하는 계층**</font>
- **구현 장치**: L5 스위치
- **하는 일**: 동시 송수신(duplex), 반이중 방식(half-duplex), 전이중 방식(full duplex) 통신을 통한 사용자 동기화 및 체크 포인팅, 유휴, 종료, 재시작 등을 수행


### -4계층: 전송 계층(Transport Layer)
- <font color=blue>**양 끝단의 사용자들이 신뢰성 있는 데이터를 주고받을 수 있게 해주는 계층**</font>
- **전송 단위**: Segment(data + TCP header: header에는 애플리케이션의 포트 번호가 있음)
- **구현 장치**: L4 스위치
- **프로토콜**: TCP, UDP
- **하는 일**: 패킷들의 전송이 유효한지 확인(오류검출/복구/흐름제어/중복검사) 후 전송 실패한 패킷을 재전송


### -3계층: 네트워크 계층(Network Layer)
- <font color=blue>**목표 노드까지 찾아가도록 네트워크 경로를 찾아주는 계층**</font>
- **전송 단위**: Packet
- **구현 장치**: 라우터, L3 스위치
- **프로토콜**: IP, ICMP, ARP/NDP, RIP, OSPF, IGRP, BGP, ...
- **하는 일**: 목표 노드를 찾아가는 과정에서 요구되는 서비스 품질(**QoS**)을 제공하기 위해, **라우팅, 패킷 포워딩, 세그맨테이션, 인터네트워킹 등을 수행**  
  ※여러 대의 노드를 거치는 복잡한 네트워크를 구성하려면 3계층 이상이 필수적임


### -2계층: 데이터링크 계층(DataLink Layer)
- <font color=blue>**하드웨어 간의 네트워크 구성 방식에 대한 계층**</font>
- **전송 단위**: frame
- **구현 장치**: 브릿지, L2 스위치 (**물리적 주소(MAC Adress)**를 통해 데이터를 정해진 곳으로 전달)
- **구성 방식 종류**: 이더넷, USB, Bluetooth, WiFi, HDLC, ADCCP, LLC, ALOHA, ...
- **하는 일**: 점대점(P2P) 연결 간 신뢰성을 보장하기 위해, 순환 중복 검사(CRC) 기반의 오류 제어, 흐름제어 수행  
  ※반도체 설계 단계에서 구현되며(즉, **하드웨어**임), 보통 1,2계층을 통틀어 하나의 칩 형태로 만듦


### -1계층: 물리 계층(Physical Layer)
- <font color=blue>**네트워크 상에서 데이터(bit)를 전송 가능 여부를 판단하는 계층**</font> 
- **전송 단위**: bit 
- **구현 장치**: 허브, 리피터
- **하는 일**: 데이터 전송이 제대로 이뤄지는지를 확인하기 위해 장치 연결을 위한 배치, 전압, 케이블 연결여부 확인 등을 수행

 ---
 # 패킷 (Packet)
 
- <font color=blue>**네트워크에서 전달되는, 형식화된 데이터 묶음의 단위**</font>
- 원래는 우편 용어로, Package(화물) + Bucket(덩어리)의 합성어
- 큰 데이터를 한번에 보낼 경우 대역폭이 커서 느려질 수 있으므로, 패킷 단위로 나누어 전송 (**패킷 교환 방식**)

---
# 패킷의 구조 (Structure of Packet)

- **버전**: 사용 중인 버전을 식별하기 위한 데이터 / 버전 4(IPv4)와 버전 6(IPv6)가 있음
- **TTL(Time to Live)**: 패킷이 네트워크 상에 잔류할 수 있는 시간에 대한 데이터 / 해당 패킷을 처리하는 라우터는 TTL을 하나 이상씩 감소시킴
- **프로토콜**: 데이터를 저장한 방식(통신규약)을 저장한 데이터 / TCP(Transmission Control Protocol)과 UDP(User Datagram Protocol)이 있음
- **헤더 체크섬**: 패킷 내 데이터의 손상여부를 확인하기 위한 데이터
- **소스 주소**: 패킷을 네트워크로 보낸 장치의 IP 주소 / ICMP(Internet Control Message Protocol) 실패, 제어 메시지, 네트워크 수준 오류도 여기로 전송
- **대상주소**: 패킷이 도착할 장치의 IP 주소
- **데이터**: 전달하고자 하는 본 데이터

# 라우팅 (Routing)
- <font color=blue>**네트워크 상의 주소를 이용하여 데이터를 목적지까지 보낼 최적 경로를 선택하는 과정**</font>
- 라우팅 = 경로배정 = 경로제어
- 라우터(Router): 라우팅을 능동적으로 수행하는 장치
- 포워딩 테이블(Forwarding Table): 입력 포트의 수신 패킷을 어느 출력 포트로 보낼 것인지를 나타낸 표 
- 라우팅 테이블(Routing Table): 포워딩 테이블 및 최적 라우팅 정보를 모두 나타낸 표 / 목적지, 방향, 거리 정보가 담김
- 라우팅 메트릭(Routing Metric): 목적지에 이르는 여러 경로에 대한 비용을 숫자로 환산한 값(Path Length, Reliability, Delay, Bandwidth, Load, Cost)
- **정확성(correctness), 단순성(simplicity), 견고성(robustness), 안정성(stability), 공정성(fairness), 최적성(optimality)** 등의 요구사항이 필요


### 경로 선택 방식에 따른 주요 구분
- **정적 라우팅(Static Routing)**
  - 관리자가 고정적인 형태로 라우팅 테이블을 작성하는 기법(수동 경로 설정) = 고정 라우팅(Fixed Routing)
  - 경로 정보의 갱신이 없으므로, 네트워크 변화에 취약함  
  - **최단경로 라우팅(Shortest Path Routing)**: 
  - **플러드 라우팅(Flood Routing / Flooding)**: 수신 경로 이외의 모든 경로로 전파
  - **랜덤 라우팅(Random Routing)**: 무작위 경로로 전파


- **동적 라우팅(Dynamic Routing)**
  - 라우터끼리 정보를 주고받아서 라이팅 테이블을 작성하는 기법(자동 경로 설정) = 적응형 라우팅(Adaptive Routing)
  - 경로정보의 수집/관리로 인해 성능 저하 발생  
  - **거리 벡터 라우팅**
  - **링크 상태 라우팅**
  
### 경로 선택 주체에 따른 주요 구분
- **글로벌 라우팅(Global Routing)**
  - 네트워크 전체의 정보를 가지고 최소 비용 경로를 계산/선택
  
  
- **분산 라우팅(Decentralized Routing)**
  - 각 라우터들 간 반복적이고 분산된 정보 전송을 통해 최소 비용 경로를 계산/선택

### 기타 라우팅
- **소스 라우팅(Source Routing)**: 경로를 모두 리스트로 만들어 패킷의 헤더에 넣어 보내는방식
- **계층적 라우팅(Hierarchical Routing)**: 큰 그룹에서 작은 그룹으로 경로를 선택하는 방식

# 라우팅 프로토콜 (Routing Protocol)
- 라우터 간 라우팅 정보의 교환 및 라우팅 테이블의 관리를 동적으로 수행하는 프로토콜



### 라우팅 프로토콜의 요소
- **라우팅 수행 프로세스**: 라우팅 알고리즘 수행/ 토폴로지 변화를 감지하고 자동으로 학습 및 반영
- **라우팅 알고리즘**: 최적 경로를 산출하면서 라우팅 테이블의 갱신 및 유지관리
- **라우팅 프로토콜 메시지**: 라우팅 정보를 다른 라우터들로 운반/전달/교환


### 라우팅 프로토콜의 구분
#### 라우터에 전달되는 정보의 종류에 따른 구분
- 링크 상태(Link State/ **LS**)   
  - 각 라우터는 네트워크 전체의 상황을 알고, 이를 토대로 경로를 선택
  - 대표적인 알고리즘으로, **다익스트라 알고리즘(Dijkstra Algorithm)**이 있음
  - 대표적인 방식으로, **OSPF, IS-IS**가 있음
  - 이벤트 기반의 경신 신호 교환 (컨버전스 시간이 작음)
  - 링크상태 정보만을 교환
  - 대규모 네트워크에 적합
    
    
- 거리 벡터(Distance Vector/ **DV**)
  - 각 노드는 연결된 이웃노드들이 보내는 정보만을 알고, 이를 토대로 경로를 선택
  - 대표적인 알고리즘으로, **벨만-포드 알고리즘(Bellman-Ford Algorithm)**이 있음
  - 대표적인 방식으로, **RIP, IGRP, EIGRP**가 있음
  - 주기적으로 경신 데이터를 교환 (컨버전스 시간이 큼)
  - 이웃한 라우터와 라우팅 테이블 교환
  - 소규모 네트워크에 적합
  
  
- DV의 특징  
  - 이웃의 정보를 토대로 계산 후 이웃에게 계산 결과를 전달함 (**분산적**)
  - 이웃끼리 라우팅 정보를 반복적으로 주고받음 (**반복적**)
  - 이웃에서 보내준 경로 예측값들과 자신의 경로 예측값을 비교하여, 적은 값을 최적경로로 삼음 (**비동기적**)
  
※ **컨버전스 시간(convergence time)**: 라우터 간에 서로 변경된 정보를 주고받는데 걸리는 시간  
※라우팅 프로토콜 별 각각 고유한 메트릭 값을 지님
  - RIP: Hop Count
  - OSPF: Link Cost
  - IGRP, EIGRP: Bandwidth, Delay, Load, Reliability, MTU 등 여럿을 조합하여 계산한 값이 가장 낮은 경로
  - BGP: 네트워크 도달성(Reachability) 정보


#### AS (Autonomous System) 내부간(Intra-domain), 외부간(Inter-domain)에 따른 구분
- AS Intra-domain: IGP(Interior Gateway Protocol)
    - RIP, OSPF, IS-IS
    
    
- AS Inter-domain: EGP(Exterior Gateway Protocol)
    - EGP, BFP, IDRP

# 애드혹 네트워크 (Ad-Hoc network)
- <font color=blue>**독립된 노드끼리 외부의 도움없이 스스로 자율적인 임시적 망을 구성하는 것**</font> (즉, 고정된 유선망 및 중앙제어가 없음)
- 임시적 망이 이동(Mobile)하는 경우, **Mobile Ad-Hoc NetWork(MANET)** 이라고 함
- 이동이 자유롭기 때문에 토폴로지가 계속 변화함 (**동적 토폴로지**)
- 단말의 이동에 제약이 없으며, 기지국(Base Station)이 필요없음
- 군사적 목적에서 출발한 개념: 고립상황에서 이동 단말들이 어떻게 네트워크를 구축해 통신할 수 있을지

#### Ad-Hoc Network와 반대되는 개념: <font color=blue>Infrastructure Network</font>
기지국을 기점으로 한 고정된 유선망으로, 이동전화망, 무선 LAN 등이 있음


<center>MANET의 단말 중 하나라도 인터넷에 연결되면, 전체 망이 인터넷에 접속할 수 있게 됨</center>

# 인터넷 (Internet)

- <font color=blue>**인터넷 프로토콜 스위트(TCP/IP)를 기반으로 한 광역 네트워크**</font>
- Inter Network의 줄임말
- 미 국방부 고등 연구국(ARPA/현재의 DARPA)에서 만든 네트워크인 **ARPANET**을 민간용으로 확장시킨 것 (현재 미군은 MILNET을 만들어 사용)
- World Wide Web(W3) 외에도, 전자 메일, 토렌트, 웹캠, 온라인 게임, 모바일 앱 등 수많은 종류의 인터넷이 존재

# 인트라넷 (Intranet)

- 인터넷에서 사용하는 회선 및 기술을 이용하여 구축한 <font color=blue>**사설 네트워크**</font>
- 주로 군대, 기업 내부에서 보안을 위해 사용
- 인터넷이 생긴 이후로 이와 구분하기 위해 생긴 단어 / ARPANET은 인트라넷이므로, 순서상 인터넷보다 인트라넷이 먼저 생김
- 외부에서 승인된 사람들도 접속이 가능한 경우, 엑스트라넷(Extranet)이라고 함 (상호 인트라넷을 연결)

# 와이파이 (Wi-Fi)

- 무선 통신 표준 기술 중 하나인 IEEE 802.11에 기반한, <font color=blue>**장치간 데이터 전송 규약**</font>
- <font color=red>**데이터 전송 규약일 뿐이므로, 인터넷 접속과 상관없음**</font>
- High-Fidelity의 약자인 Hi-Fi를 Wi-Fi로 바꾼 것으로, 고유명사이며, Wi-Fi Alliance에 가입된 회사가 쓸 수 있는 **상표명**
- WLAN (Wireless LAN)이라고 하기도 함 (이 경우는 Wi-Fi Alliance에서 인증을 받지 못한 경우 사용)
- 무선이므로, AP가 필요 (**AP: Access Point / 소형 무선 단말기**)
- **일대다 통신 방식(infrastructure mode)**을 사용하기 때문에, <font color=red>**연결된 장치가 증가하면 전송 속도가 반비례하여 감소**</font>
- 근거리 통신을 전제로 하기 때문에, 개활지에서 **반경 200m** 정도가 사용 가능 범위  
※유선 프로토콜인 이더넷 또한 근거리 통신을 전제로 하므로, UTP 케이블을 100m 이상 사용하지 않는걸 권장함 (길어질수록 느려짐)

# TCP (Transmission Control Protocol)

- **OSI 4 layer(Transport) 프로토콜**의 일종
- **연결형 서비스**: client와 server가 연결된 상태에서 데이터를 주고받는 방식(**가상 회선 방식**)
- **바이트 스트림 서비스**: 데이터의 경계를 구분하지 않음
- **데이터의 전송순서 보장**: 데이터의 순서 유지를 위해 각 바이트마다 번호 부여
- **신뢰성있는 데이터 전송**: Sequence Number, ACK Number를 통한 신뢰성 보장
- **데이터 흐름 제어 및 혼잡 제어**: 수신자 버버 오버플로우 방지 및 패킷 수가 과도하게 증가하는 현상 방지
- **3 way handshake**: client가 연결을 요청하며 SYN 전송 -> server가 연결을 수락하고 SYN ACK 전송 -> client가 전송성공 후 ACK 전송 / 실패 시 재시도
- <font color=red>**단점**: 데이터를 보내기 전 반드시 연결이 생성되어 있어야 함 / 1:1 통신만 가능 / UDP보다 상대적으로 느림</font>  


### ※TCP state
- **LISTEN** : server의 데몬이 떠서 접속 요청을 기다리는 상태
- **SYN-SENT** : 로컬의 client 어플리케이션이 원격 호스트에 연결을 요청한 상태
- **SYN_RECEIVED** : server가 원격 클라이언트로부터 접속 요구를 받아 client에게 응답을 하였지만 아직 client에게 확인 메시지는 받지 않은 상태
- **ESTABLISHED** : 3 way-handshaking 이 완료된 후 서로 연결된 상태
- **FIN-WAIT1, CLOSE-WAIT, FIN-WAIT2** : server에서 연결을 종료하기 위해 client에게 종결을 요청하고 회신을 받아 종료하는 과정의 상태
- **TIME-WAIT** : 연결은 종료되었지만 분실되었을지 모를 느린 세그먼트를 위해 당분간 소켓을 열어두고 있는 상태
- **CLOSING** : 흔하지 않지만 주로 확인 메시지가 전송도중 분실된 상태 
- **CLOSED** : 완전히 종료

# UDP (User Datagram Protocol)

- **OSI 4 layer(Transport) 프로토콜**의 일종
- **비연결형 서비스**: client와 server가 연결 절차 없이, server가 일방적으로 데이터를 송신하는 방식(**데이터그램 방식**)
- **데이터그램 서비스**: 데이터의 경계를 구분함
- **데이터의 전송순서 미보장**: 데이터의 순서 유지를 위한 작업을 하지 않음
- **패킷 오버헤드 감소**: 정보를 주고받는다는 신호절차 및 순서 유지 작업을 하지 않기 때문에, 네트워크의 부하가 감소 => TCP보다 상대적으로 빠름
- <font color=red>**단점**: 데이터의 신뢰성이 없음 / 의미있는 서버 구축을 위해선 일일이 패킷을 관리해줘야 함</font>
